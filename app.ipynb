```python
# web = requests.get("https://www.tutorialsfreak.com/")
# web = requests.get('https://www.crummy.com/')
web = requests.get('https://abidkk.site/')
web
```

# 2. Data Parsing 

```python
soup = BeautifulSoup(web.content, 'html.parser')
print(soup.prettify())
```

# Tags

```python
print(soup.title)
print(soup.h1)
print(soup.p)
print(soup.a)
print(soup.img)
```

# String

```python
print(soup.p.string)
print(soup.title.string)
```

# Beautifulsoup

```python
print(soup.name)
print(soup.title)
print(soup.head)
print('\n --- \n ')
print(soup.body)
```

# Beautifulsoup : Function

- find('p') : find first p tag

```python
soup.find('p')
```

- findall('h1')

```python
soup.find_all('p')
```

# Comments

```python
single_comment = soup.find(string=lambda text: isinstance(text, Comment))

# Print the single comment
if single_comment:
    print(single_comment)
else:
    print("No comments found.")
```

```python
# Find all comments
comments = soup.find_all(string=lambda text: isinstance(text, Comment))

# Print the comments
for comment in comments:
    print(comment)
```
